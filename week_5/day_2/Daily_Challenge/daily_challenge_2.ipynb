Generate a Seaborn bar plot to compare the number of movies across different genres

```python
genre_group = df.groupby('Genre').count()
df['Genre'].unique #there are many genre combinations. I should split them by individual genres.
```

```python
genre_list = df['Genre'].str.split(', ') #I want to isolate each individual genre in each Genre field w/ multiple genres

new_df = df.assign(Genre=genre_list) # assigning list to a new df
new_df = new_df.explode('Genre').reset_index(drop=True) # breaking each item in the list into its own row
new_df
```

```python
genre_group = new_df.groupby('Genre').count()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=genre_group.index,y=genre_group['Director'],palette='pastel')
plt.title('Directors per genre')
plt.xticks(rotation=45)
plt.xlabel('Genre',fontweight='bold')
plt.ylabel('Director',fontweight='bold')
plt.show()
```

Use a Matplotlib scatter plot to examine the relationship between directors and movie ratings.

```python
# identify all directors w/ more than 3 movies

director_counts = df['Director'].value_counts()
directors_more_than_twice = director_counts[director_counts > 3]
directors_list = list(directors_more_than_twice.index)
directors_list
```

```python
limited_df = df[df['Director'].isin(directors_list)] # only directors w/ many films
limited_df.groupby('Director').count()
```

```python
# sns.lmplot(x=director_group.index, y=director_group['IMDB_Rating'], data=director_group)
plt.figure(figsize=(23,8))
plt.scatter(limited_df['Director'], limited_df['IMDB_Rating'],s=30)
plt.xticks(rotation=70)
plt.xlabel('Director',fontweight='bold')
plt.ylabel('IMDB Rating',fontweight='bold')
plt.title('Rating distribution by director',fontweight='bold')
plt.show()
```

Create a heatmap using Seaborn to identify correlations between numerical columns like IMDB_Rating, Meta_score, No_of_votes, and Gross.

```python
df_cleaned = df[['Meta_score','IMDB_Rating','Gross','No_of_Votes']] # cleaning gross column
len(df_cleaned)
```

```python
matrix = df_cleaned['Gross'].notna()
df_cleaned = df_cleaned[matrix]
```

```python
df_cleaned
```

```python
correlation_matrix = df_cleaned.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
