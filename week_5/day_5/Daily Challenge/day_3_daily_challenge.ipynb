```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset('iris')

df = pd.DataFrame(data)
```

Answer: Based on the below scatterplot, there is a weak negative correlation between sepal length and width overall but when you break it down by specie, you find a positive correlation in the case of virginica and and versicolor. It's a much sharper positive correlation for setosa.

```python
plt.figure(figsize=(6,8))

sns.lmplot(x='sepal_length',y='sepal_width', data=data)
plt.title('Sepal length to sepal width ratio')
plt.show()
```

```python
plt.figure(figsize=(6,8))

sns.lmplot(x='sepal_length',y='sepal_width', data=data, hue='species')
plt.title('Sepal length to sepal width ratio')
plt.show()
```

```python
plt.hist(df['petal_length'],bins=20)
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.title('Petal Length count')
plt.show()
```

```python
sns.pairplot(data, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
Create an interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species. Each plot should be well-labeled with appropriate titles and axis labels.
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(data, x='sepal_length', y='sepal_width', z='petal_length', color='species', symbol='species',
                    opacity=0.7, size_max=10)
fig.update_layout(title='Sepal Length, Sepal Width, and Petal Length', scene=dict(xaxis_title='Sepal Length',
                  yaxis_title='Sepal Width', zaxis_title='Petal Length'))

fig.show()
```
