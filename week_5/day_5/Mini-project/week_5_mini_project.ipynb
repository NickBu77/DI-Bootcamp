Flipper Length vs. Body Mass: Create a scatter plot to analyze the relationship between flipper length and body mass, color-coded by species.


Answer: Based on the scatter plot below, all of the species have more or less linear postive relationship between their body mass and flipper length.  Histograms of Measurements: Create histograms to show the distribution of key measurements (bill length, bill depth, flipper length).

Insightful Visualizations:

Combine multiple plots into a single figure (using subplots) to compare different aspects of the data.

Based on the KDEs below, none of the species have a conventional normal distribution. Flipper length peaks further left and bill depth peaks further right, but all have dips in the middle. This may be due to sex differences or evolutionary pressures that allow large individuals to survive through elimination of rivals, allowing only less threatening individuals to survive. Worthy of further analysis.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Creating a 1x3 grid for the plots

sns.histplot(df['flipper_length_mm'], bins=10, color='purple', ax=axes[0])
axes[0].set_xlabel('flipper_length_mm')

sns.histplot(df['bill_depth_mm'], bins=10, color='skyblue', ax=axes[1])
axes[1].set_xlabel('bill_depth_mm')

sns.histplot(df['bill_length_mm'], bins=10, color='pink', ax=axes[2])
axes[2].set_xlabel('bill_length_mm')

plt.suptitle('Distribution of measurements')
plt.show()


```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Creating a 1x3 grid for the plots

sns.kdeplot(df['flipper_length_mm'], fill=True, color='purple', ax=axes[0])
axes[0].set_xlabel('flipper_length_mm')

sns.kdeplot(df['bill_depth_mm'], fill=True, color='skyblue', ax=axes[1])
axes[1].set_xlabel('bill_depth_mm')

sns.kdeplot(df['bill_length_mm'], fill=True, color='pink', ax=axes[2])
axes[2].set_xlabel('bill_length_mm')

plt.suptitle('Distribution of measurements')
plt.show()
```
