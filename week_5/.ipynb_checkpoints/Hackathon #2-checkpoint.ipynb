```python
path = r'C:\Users\97258\Downloads\Foodex_2_L7_dashboard.xlsx'

data = pd.read_excel(path, skiprows=2)
EU_df = pd.DataFrame(data)
EU_df
```

```python
most_recent_yr = EU_df.groupby('Survey\'s country')['Survey start year'].max() # Because there are multiple years in which survey were conducted, I wanted to take the most recent year only.
most_recent_yr
```

```python
year_df = EU_df[EU_df.apply(lambda x: x['Survey start year'] == most_recent_yr[x['Survey\'s country']], axis=1)] #function to pull only max year from original df
year_df
```

```python
# Here I am just doublebecking that the year in the newly created year_df really matches the most_recent_year grouping from before. 
years = year_df.groupby('Survey\'s country')['Survey start year'].mean()

check = pd.merge(most_recent_yr, years, on='Survey\'s country')
mismatches = check['Survey start year_x'] != check['Survey start year_y']
#check[~mismatches]
print(f'There are {len(check[mismatches])} mismatches')
```
