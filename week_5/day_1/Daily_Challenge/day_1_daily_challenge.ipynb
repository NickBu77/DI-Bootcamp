```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\97258\Documents\DI-Bootcamp\week_5\Daily_Challenge\world_happiness_index\2019.csv')
df = pd.DataFrame(data)
df
```

```python
scores = list(df['Score'])
social_support = list(df['Social support'])


plt.scatter(social_support, scores, color='red')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.title('Social support to happiness')
```
