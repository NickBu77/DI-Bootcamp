🌟 Exercise 2 : Scraping Robots.Txt From Wikipedia
Instructions
Write a Python program to download and display the content of robot.txt for en.wikipedia.org

```python
from urllib.request import urlopen
url = 'https://en.wikipedia.org/robots.txt'
response = urlopen(url)
html = response.read()
soup = BeautifulSoup(html, 'html.parser')
soup.prettify()
```



🌟 Exercise 3 : Extracting Headers From Wikipedia’s Main Page
Instructions
Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page.


```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
response = urlopen(url)
html = response.read()
soup = BeautifulSoup(html, 'html.parser')
soup.prettify()
```

```python
soup.h1
```

```python
headers= []

headers.append(soup.find_all("h1"))
headers.append(soup.find_all("h2"))
headers.append(soup.find_all("h3"))
headers.append(soup.find_all("h4"))
headers.append(soup.find_all("h5"))
headers.append(soup.find_all("h6"))

headers
```

```python
cleaned_headers_list = headers[:-4]
cleaned_headers_list
```
