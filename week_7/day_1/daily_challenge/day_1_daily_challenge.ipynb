```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://github.com/topics'
response = urlopen(url)
html = response.read().decode('utf-8')
html
```

```python
with open('webpage.html','w', encoding='utf-8') as file:
    file.write(html)
```

```python
with open('webpage.html', 'r') as f:
    open_html = f.read()
    
soup = BeautifulSoup(open_html, 'html.parser') 
soup
```

```python
courses = soup.find_all(class_="f3")
len(courses)
```

```python
descriptions = soup.find_all('p',class_="f5")
len(descriptions)
```

```python
df_desc = [desc.get_text(strip=True) for desc in descriptions]
df_course = [course.get_text(strip=True) for course in courses]
```

```python
df_dict = {'title': list(course for course in df_course)}
df_dict['description'] = list([desc for desc in df_desc])
print(df_dict)
```

```python
import pandas as pd

df = pd.DataFrame(df_dict)
df
```
