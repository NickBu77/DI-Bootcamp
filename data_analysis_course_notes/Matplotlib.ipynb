### Basics of Matplotlib: Figures and Axes

A Matplotlib visualization is a figure that contains one or more plots, also called "axes". You can think of the "figure" as the whole window or page, and each "axes" as an individual plot or graph within that window. Each axes has an x (horizontal) and y (vertical) axis, and we represent data using color and glyphs (symbols like circles, lines, or polygons).


````python
# Creating a figure and axes
fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes
````

````python
# Creating a figure and axes
fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.plot([10, 12, 14, 16], [15, 17, 19, 16])  # Plot some data on the axes
````

## Creating and Customizing Basic Plots in Matplotlib

### Plotting Basics - Figure and Axes

Matplotlib provides a Figure object, which is the overall window or page that everything is drawn on, and an Axes object, which is a bounding box with ticks and labels.

Creating a new figure and axes is as simple as


```
fig, ax = plt.subplots()
```







In this case, 'fig' is a Figure object and 'ax' is an Axes object.

Once you have an Axes object, you can call a method to draw on it:

````python
fig, ax = plt.subplots()  # Create a figure and an axes
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes
````


### Basic Plots

Matplotlib provides a variety of basic plots. Here are some examples:

1. **Line Plot**

   A line plot is generally used to present observations collected at regular intervals. The x-axis represents the regular interval, such as time. The y-axis shows the observations, ordered by the x-axis and connected by a line.


````python
import numpy as np

x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced numbers between 0 and 10
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
````

2. **Scatter Plot**

  A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point.

````python
x = np.random.rand(50)
y = np.random.rand(50)

# x = [1,2,3,4,5,2,34,0]
# y = [2,12,4,78,4,5,1, 0.45]
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
````

````python
ages = [15, 24, 37, 18, 42, 59, 70 ]
incomes = [1, 14, 22, 6, 12, 28, 11]
````

````python
fig, ax = plt.subplots()
ax.scatter(ages, incomes)
plt.show()
````

### Styling Plots

There are many ways to style your plots in matplotlib to make them look more appealing.

- **Title, Labels:** You can add a title to your plot and label the x and y axes using the `set()` function:

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax.set(title='A Simple Plot', xlabel='x', ylabel='y')
plt.show()
````

We'll look at some more styling options later.

Let's redo the previous age - income scatter plot but this time add titles.

````python
fig, ax = plt.subplots()
ax.scatter(ages, incomes)
ax.set(title='Incomes By Age', xlabel='Age', ylabel='Income(k)/Month')
plt.show()
````



3. **Histogram**

   A histogram takes in a series of data and divides the data into a number of bins. A bin refers to the range of values that is divided into series of **continuous** intervals. It then plots the frequency of data points in each bin (i.e., how many values fall in each interval).

````python
data = np.random.randn(400)

fig, ax = plt.subplots()
ax.hist(data, bins=50)
plt.show()
````

````python
mydata = [1,2,3,4,1,2,5,4,1,2,3,1,3,4,5,3,2,1]
fig, ax = plt.subplots()
ax.hist(mydata, bins=10)
plt.show()
````

### Saving Plots

You can save the figures you create using the `savefig()` function. This will save the figure as a PNG file. You can also save it as PDF, SVG, JPG, and others. The format is inferred from the extension you provide in the filename.

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
fig.savefig('my_new_figure.png')  # save the figure to file
````


## Common Pitfalls

1. Don't forget to call `plt.show()` after plotting. Without this call, your plots might not render or display.
2. Ensure that the data types you're plotting are compatible with the plot type you're using. For instance, string values cannot be plotted on a histogram.
