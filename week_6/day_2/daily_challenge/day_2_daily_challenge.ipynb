Perform a statistical analysis of power output by fuel type using NumPyâ€™s statistical functions. Use hypothesis testing to determine if the mean power output differs significantly between different fuel types.

```python
avg = df.groupby('primary_fuel')['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015',
       'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018',
       'generation_gwh_2019'].mean()

avg
```

```python
avg['fuel_sum'] = avg.mean(axis=1)
wnt = avg.index=='Wave and Tidal'
avg = avg[~wnt]
```

```python
avg
```

```python
maxvalue= avg['fuel_sum'].max()
maxindex = avg['fuel_sum'].idxmax()
minvalue= avg['fuel_sum'].min()
minindex = avg['fuel_sum'].idxmin()

print(f'The max is {maxvalue}, by energy source {maxindex}')

print(f'The min is {minvalue}, by energy source {minindex}')

print('This is a huge discrepancy. Solar is highly inneficient relative to nuclear.')
```
