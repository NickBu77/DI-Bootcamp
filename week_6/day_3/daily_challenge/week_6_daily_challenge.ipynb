Statistical Analysis:

Apply SciPy to analyze the distribution of fatalities and survival rates. Calculate key statistics like mean, median, and standard deviation.

Conduct a hypothesis test (e.g., comparing the average number of fatalities in different decades or regions).

Answer: My null hypthesis is that there was no significant difference in total fatalities and in fatality rates across decades.

My alternative hypothesis is that the total fatalities would go down over time and the fatality rate would decrease over time because of improved technology that help w/ crash landings, as well as imporived communication and medical care that could quickly retrieve and help save people.

```python
no_nans['Year'] = no_nans['Date'].dt.year
```

```python
no_nans['Decade'] = np.nan
```

```python
for index, row in no_nans.iterrows():
  if row['Year']  >= 2020:
    no_nans.loc[index,'Decade']='2020+'
  elif row['Year'] >= 2010:
    no_nans.loc[index,'Decade']='2010-2019'
  elif row['Year'] >= 2000:
    no_nans.loc[index,'Decade']='2000-2009'
  elif row['Year'] >= 1990:
    no_nans.loc[index,'Decade']='1990-1999'
  elif row['Year'] >= 1980:
    no_nans.loc[index,'Decade']='1980-1989'
  elif row['Year'] >= 1970:
    no_nans.loc[index,'Decade']='1970-1979'
  elif row['Year'] >= 1960:
    no_nans.loc[index,'Decade']='1960-1969'
  elif row['Year'] >= 1950:
    no_nans.loc[index,'Decade']='1950-1959'
  elif row['Year'] >= 1940:
    no_nans.loc[index,'Decade']= '1940-1949'
  elif row['Year'] >= 1930:
    no_nans.loc[index,'Decade']='1930-1939'
  elif row['Year'] >= 1920:
    no_nans.loc[index,'Decade']='1920-1929'
  elif row['Year'] >= 1910:
    no_nans.loc[index,'Decade']='1910-1919'
  else:
    no_nans.loc[index,'Decade']='Before 1910'
```

```python
group = no_nans.groupby('Decade')['Fatalities','fatality_rate'].mean()
group = group.reset_index()
```

```python
group
```

```python
index_list = list(group.index)
index_list.insert(0, index_list.pop(12))

index_list
```

```python
reordered_group = group.loc[index_list]
```

```python
reordered_group
```

```python
fig, ax1 = plt.subplots(figsize=(15,6))

x = reordered_group['Decade']
y = reordered_group['fatality_rate']
y2 = reordered_group['Fatalities']

ax1.plot(x,y, color='r')
ax1.set_ylabel('Fatality Rate', color='r')
ax1.tick_params(axis='y', colors='r')

ax2 = ax1.twinx()
ax2.bar(x,y2, alpha=.5)
ax2.set_ylabel('Fatalities', color='b')
ax2.tick_params(axis='y', colors='b')


plt.xticks(rotation=50)

plt.xticks(rotation=45) #why doesn't this work?
plt.title('Fatalities and fatality rate')
plt.show()
```

The first real decade of data, 1910-1919, had a casualty rate near 1. However, it drops pretty quickly and continues to do so until the 1990's, at which point it begins to rise again.
