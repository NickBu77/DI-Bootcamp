```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(0)

# Number of entries
n_entries = 1000

# Simulating dataset
time = np.linspace(0, 100, n_entries)  # Time variable
temperature = 20 + 5 * np.sin(np.pi * time / 50) + np.random.normal(0, 0.5, n_entries)  # Temperature (with some noise)
pressure = 1013 + 20 * np.cos(np.pi * time / 25) + np.random.normal(0, 1, n_entries)  # Pressure (with some noise)
chemical_concentration = 5 + 2 * np.sin(np.pi * time / 10) + np.random.normal(0, 0.2, n_entries)  # Chemical concentration

# Creating DataFrame
ninja_data = pd.DataFrame({
    'Time': time,
    'Temperature': temperature,
    'Pressure': pressure,
    'Chemical Concentration': chemical_concentration
})
```

```python
ninja_data
```

```python
import pandas as pd
import numpy as np
from scipy import stats

normalized_ninja = stats.zscore(ninja_data)
norm_df = pd.DataFrame(normalized_ninja)
norm_df
```

```python

norm_df.columns = ninja_data.columns
```

```python
mean = stats.tmean(norm_df['Time'])
med = np.median(norm_df['Time'])
std = stats.tstd(norm_df['Time'])
min = stats.tmin(norm_df['Time'])
max = stats.tmax(norm_df['Time'])


f'Mean: {mean}, Median: {med}, Stand Dev: {std}, Max: {max}, Min: {min}'
```

```python
mean = stats.tmean(norm_df['Temperature'])
med = np.median(norm_df['Temperature'])
std = stats.tstd(norm_df['Temperature'])
min = stats.tmin(norm_df['Temperature'])
max = stats.tmax(norm_df['Temperature'])


f'Mean: {mean}, Median: {med}, Stand Dev: {std}, Max: {max}, Min: {min}'
```

```python
mean = stats.tmean(norm_df['Pressure'])
med = np.median(norm_df['Pressure'])
std = stats.tstd(norm_df['Pressure'])
min = stats.tmin(norm_df['Pressure'])
max = stats.tmax(norm_df['Pressure'])


f'Mean: {mean}, Median: {med}, Stand Dev: {std}, Max: {max}, Min: {min}'
```

```python
mean = stats.tmean(norm_df['Chemical Concentration'])
med = np.median(norm_df['Chemical Concentration'])
std = stats.tstd(norm_df['Chemical Concentration'])
min = stats.tmin(norm_df['Chemical Concentration'])
max = stats.tmax(norm_df['Chemical Concentration'])


f'Mean: {mean}, Median: {med}, Stand Dev: {std}, Max: {max}, Min: {min}'
```

```python
norm_df.describe()
```

```python
corr_matrix = norm_df.corr()
corr_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap", fontsize=16, fontweight='bold') 

plt.show()
```

```python
sns.pairplot(norm_df, kind='scatter', plot_kws={'s':3})
plt.suptitle('Time vs Temp vs Pressure vs Chem Con', y =1.02)
plt.show()
```

```python
t_stat, p_value = stats.ttest_ind(norm_df['Time'], norm_df['Temperature'])
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
t_stat, p_value = stats.ttest_ind(norm_df['Time'], norm_df['Pressure'])
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
t_stat, p_value = stats.ttest_ind(norm_df['Time'], norm_df['Chemical Concentration'])
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
f_value, p_value = stats.f_oneway(norm_df['Time'],norm_df['Temperature'], norm_df['Pressure'], norm_df['Chemical Concentration'])
print("F-value:", f_value)
print("P-value:", p_value)
```

```python
pairplot = sns.pairplot(norm_df, kind='scatter', plot_kws={'s': 3})
plt.suptitle('Time vs Temp vs Pressure vs Chem Conc', y =1.02)
plt.show()
```

Because the data was normalized, it appears that the t-stats and f-value will produce values close to 0. The heatmap above also showed no significant correlation with the exception of time and temperature, which have a strong negative correlation of -.77. This may be misleading though. 

The only graphs of value really are the scatterplots, which show clear but non-linear patterns. Time vs temp has a sideways 'S' pattern. As time begins to increase, temperature increases, then it sees a steady and long decline before rising again at the end of the timescale. The temp at time 0 is aligned with the very beginning and very end of the timescale.

The relationship of time vs pressure is similar to that of time vs temperature except that the very beginning, middle, and end of the timescale have the highest pressure measures.

The relationship of time vs chemical concentration is similar to time vs temperature, the dips and rises are just more frequent; there are more waves in other words.

Looking at the relationships of the secondary measures (all except for time) vs each other produce less clear trends. However, it appears to be the case that temperature and pressure have an arch-like relationship, peaking in the mean temperature range with both temperature extremes at their lowest pressure levels.
